## 1. Generating functions
### a)
The following function plots the generating function of a discrete univarate probability density function $p(k)$.

```python
def plot_generating_function(p: Dict[int, float], start=0, stop=1, resolution=100, label: str = ""):
    xs = np.linspace(start, stop, num=resolution)
    ys = np.zeros_like(xs)
    for i, x in enumerate(xs):
        for k, pk in p.items():
            ys[i] += pk * x**k
    plt.grid()
    plt.plot(xs,ys, label=label)


```

```python
# This allows us to generate the probability density function for a n-sided die
die = lambda n: {k + 1: 1/n for k in range(n)}
```

```python
plot_generating_function(die(6))
plot_generating_function(die(10))
plot_generating_function(die(12))
plt.legend(["$n=6$","$n=10$", "$n=12$"])
```

From the plots above, we can make several observations:
- The derivative of the generating functions at $x = 1$ increases with the number of sides $n$. We expect this since it represents the expected value of that dice which increases as we add more (consecutively numbered) sides.
- All generating functions coincide at $x = 0$ since we have $P(0) = 0$.
- All generating functions coincide at $x = 1$ since we have $\sum_{k=0}^\infty P(k) = 1$.
